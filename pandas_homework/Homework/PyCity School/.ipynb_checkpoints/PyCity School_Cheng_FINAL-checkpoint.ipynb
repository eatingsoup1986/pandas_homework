```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
totalschools = (school_data_complete["school_name"].value_counts().count())
totalstudents = school_data_complete["Student ID"].count()
totalbudget = (school_data_complete["budget"].unique()).sum()
avgmath = school_data_complete["math_score"].mean()
avgreading = school_data_complete["reading_score"].mean()
percentmathpass = ((school_data_complete["math_score"]>=70).sum()/totalstudents)*100
percentreadpass = ((school_data_complete["reading_score"]>=70).sum()/totalstudents)*100
percenttotalpass = (((school_data_complete["math_score"]>=70) & ((school_data_complete["reading_score"]>=70))).sum()/totalstudents)*100

#display summary in dataframe
summary_df = pd.DataFrame(
                {"Total Schools":[totalschools],
                "Total Students":[totalstudents],
                "Total Budget":[totalbudget],
                "Average Math Score":[avgmath],
                "Average Reading Score":[avgreading],
                "% Passing Math":[percentmathpass],
                "% Passing Reading":[percentreadpass],
                "% Overall Passing":[percenttotalpass]})

#set formatting for appropriate cells
summary_df["Total Budget"] = summary_df["Total Budget"].map("${:,.2f}".format)
summary_df["Total Students"] = summary_df["Total Students"].map("{:,}".format)
summary_df["Average Math Score"] = summary_df["Average Math Score"].map("{:.2f}".format)
summary_df["Average Reading Score"] = summary_df["Average Reading Score"].map("{:.2f}".format)
summary_df["% Passing Math"] = summary_df["% Passing Math"].map("{:.2f}%".format)
summary_df["% Passing Reading"] = summary_df["% Passing Reading"].map("{:.2f}%".format)
summary_df["% Overall Passing"] = summary_df["% Overall Passing"].map("{:.2f}%".format)

summary_df
```

```python
#Group student data by school

school_studentcount = school_data_complete["school_name"].value_counts()
grouped_school_df = school_data_complete.groupby("school_name")
schoolBudget = grouped_school_df["budget"].max()
perstudent = schoolBudget/school_studentcount
schooltype = grouped_school_df["type"].unique()
school_avgmath = grouped_school_df["math_score"].mean()
school_avgread = grouped_school_df["reading_score"].mean()

school_student_df = pd.DataFrame(school_studentcount)

#math scores calculations for passing grade 70 or more:

school_mathpass = school_data_complete.loc[school_data_complete["math_score"]>=70]
school_mathpass_group = school_mathpass.groupby(["school_name"]).count()
school_readpass = school_data_complete.loc[school_data_complete["reading_score"]>=70]
school_readpass_group = school_readpass.groupby(["school_name"]).count()
school_percentmath = (school_mathpass_group["math_score"]/school_studentcount)*100
school_percentread = (school_readpass_group["reading_score"]/school_studentcount)*100
overall_pass = (school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('school_name')['Student ID'].count()/school_studentcount)*100
#(school_percentmath&school_percentread).count()/school_studentcount
#(school_percentmath+school_percentread)/2 #THIS ONE DOESN"T LOOK RIGHT...NEED HELP
#overall_pass = (school_readpass_group + school_mathpass_group)

#create summary dataframe
summary_school = pd.DataFrame({
    "School Type":schooltype,
    "Student Count":school_studentcount,
    "School Budget":schoolBudget,
    "Per Student Budget":perstudent,
    "Average Math Score":school_avgmath,
    "Average Reading Score":school_avgread,
    "% Passing Math":school_percentmath,
    "% Passing Reading":school_percentread,
    "% Overall Passing Rate":overall_pass
})

#clean up formating
#set formatting for appropriate cells
summary_school["School Budget"] = summary_school["School Budget"].map("${:,.2f}".format)
summary_school["Per Student Budget"] = summary_school["Per Student Budget"].map("${:,.2f}".format)
summary_school["Student Count"] = summary_school["Student Count"].map("{:,}".format)
summary_school["Average Math Score"] = summary_school["Average Math Score"].map("{:.2f}".format)
summary_school["Average Reading Score"] = summary_school["Average Reading Score"].map("{:.2f}".format)
summary_school["% Passing Math"] = summary_school["% Passing Math"].map("{:.2f}%".format)
summary_school["% Passing Reading"] = summary_school["% Passing Reading"].map("{:.2f}%".format)
summary_school["% Overall Passing Rate"] = summary_school["% Overall Passing Rate"].map("{:.2f}%".format)


summary_school
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
#group the grades together using a loc
ninthgrade = school_data_complete.loc[school_data_complete["grade"] == "9th"]
tenthgrade = school_data_complete.loc[school_data_complete["grade"] == "10th"]
elevengrade = school_data_complete.loc[school_data_complete["grade"] == "11th"]
twelvegrade = school_data_complete.loc[school_data_complete["grade"] == "12th"]

#group by school
groupninthschool = ninthgrade.groupby(["school_name"]).mean()
grouptenthschool = tenthgrade.groupby(["school_name"]).mean()
groupelevenschool = elevengrade.groupby(["school_name"]).mean()
grouptwelveschool = twelvegrade.groupby(["school_name"]).mean()

#pull together in a dataframe
mathscoregrade = pd.DataFrame({
        "9th":groupninthschool["math_score"],
        "10th":grouptenthschool["math_score"],
        "11th":groupelevenschool["math_score"],
        "12th":grouptwelveschool["math_score"]
})


mathscoregrade.head(15)
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# set up bins
spend_bins = [0, 584.99, 629.99, 644.99, 9999]
group_names = ["<$584", "$585-629", "$615-644", "$645-675"]
school_data_complete["spend_bins"] = pd.cut(school_data_complete["budget"]/school_data_complete["size"], spend_bins, labels = group_names)

# group by spend
by_spend = school_data_complete.groupby("spend_bins")

#caculations
avg_math = by_spend["math_score"].mean()
avg_read = by_spend["reading_score"].mean()
pass_math = (school_data_complete[school_data_complete['math_score'] >= 70].groupby('spend_bins')['Student ID'].count()/by_spend['Student ID'].count())*100
pass_read = (school_data_complete[school_data_complete['reading_score'] >= 70].groupby('spend_bins')['Student ID'].count()/by_spend['Student ID'].count())*100
overall = (school_data_complete[(school_data_complete['reading_score'] >= 70) & (school_data_complete['math_score'] >= 70)].groupby('spend_bins')['Student ID'].count()/by_spend['Student ID'].count())*100

#put into dataframe:
spend_scoresummary = pd.DataFrame({
        "Average Math Score": avg_math,
        "Average Reading Score": avg_read,
        "% Passing Math": pass_math,
        "% Passing Reading": pass_read,
        "% Overall Passing": overall
})
spend_scoresummary["Average Math Score"] = spend_scoresummary["Average Math Score"].map("{:,.2f}".format)
spend_scoresummary["Average Reading Score"] = spend_scoresummary["Average Reading Score"].map("{:,.2f}".format)
spend_scoresummary["% Passing Math"] = spend_scoresummary["% Passing Math"].map("{:,.2f}%".format)
spend_scoresummary["% Passing Reading"] = spend_scoresummary["% Passing Reading"].map("{:,.2f}%".format)
spend_scoresummary["% Overall Passing"] = spend_scoresummary["% Overall Passing"].map("{:,.2f}%".format)


spend_scoresummary
```
